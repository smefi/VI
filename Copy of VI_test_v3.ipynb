```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img1 = cv2.imread("0034_rgb.tiff")

#costly 
#b,g,r = cv2.split(img1)
#rbg_img = cv2.merge([r,b,g])

Blue = img1[:,:,0]
Green = img1[:,:,1]
Red = img1[:,:,2]

#plt.imshow(rbg_img)
plt.imshow(img1)
#cv2.imshow('image',img1)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
img2 = cv2.imread("0034_nir.tiff")

NIR = img2[:,:,0]
plt.imshow(NIR)
#print(img2)
#cv2.imshow('image',NIR)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
def div0( a, b ):
    """ ignore / 0, div0( [-1, 0, 1], 0 ) -> [0, 0, 0] """
    with np.errstate(divide='ignore', invalid='ignore'):
        c = np.true_divide( a, b )
        c[ ~ np.isfinite( c )] = 0  # -inf inf NaN
    return c
```

**<font color=blue size=4> EVI </font>**

```python
EVI_numerator = 2.5*(NIR-Red)
EVI_denominator = (NIR+6*Red-7.5*Blue+1)
EVI = EVI_numerator/EVI_denominator
```

```python
#print(EVI)

plt.imshow(EVI)
#cv2.imshow('image',EVI)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

**<font color=blue size=4> GLI </font>**

```python
#GLI= (Green-Red)+(Green-Blue)/(2*Green)+Red+Blue
GLI= div0(((Green-Red)+(Green-Blue)),((2*Green)+Red+Blue))
```

```python
#print(GLI)

#cv2.imshow('image',GLI)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(GLI)
```
